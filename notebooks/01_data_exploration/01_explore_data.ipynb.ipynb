```python
# Visualize Grade Distribution
import os

# Create figures directory if not exists
figures_path = r'C:\Users\Amsha\Desktop\EduMindAI\reports\figures'
os.makedirs(figures_path, exist_ok=True)

plt.figure(figsize=(12, 4))

# Plot 1 - Final Grade Distribution
plt.subplot(1, 3, 1)
sns.histplot(df_performance['G3'], bins=20, color='blue')
plt.title('Final Grade Distribution')
plt.xlabel('Grade (G3)')

# Plot 2 - Study Time vs Grade
plt.subplot(1, 3, 2)
sns.boxplot(x='studytime', y='G3', data=df_performance, color='green')
plt.title('Study Time vs Grade')

# Plot 3 - Failures vs Grade
plt.subplot(1, 3, 3)
sns.boxplot(x='failures', y='G3', data=df_performance, color='red')
plt.title('Failures vs Grade')

plt.tight_layout()
plt.savefig(os.path.join(figures_path, 'performance_analysis.png'))
plt.show()
print("✅ Chart saved!")
```

```python
# Explore Mental Health Dataset
print("="*50)
print("😰 MENTAL HEALTH DATASET")
print("="*50)
print(f"\nShape: {df_mental.shape}")
print(f"\nColumns: {df_mental.columns.tolist()}")
print(f"\nMissing Values:\n{df_mental.isnull().sum()}")
print("\nFirst 5 rows:")
df_mental.head()
```

```python
# Mental Health Visualizations
plt.figure(figsize=(12, 4))

# Plot 1 - Depression
plt.subplot(1, 3, 1)
df_mental['Do you have Depression?'].value_counts().plot(
    kind='bar', color=['green', 'red'])
plt.title('Depression')
plt.xlabel('')

# Plot 2 - Anxiety
plt.subplot(1, 3, 2)
df_mental['Do you have Anxiety?'].value_counts().plot(
    kind='bar', color=['green', 'red'])
plt.title('Anxiety')
plt.xlabel('')

# Plot 3 - Panic Attack
plt.subplot(1, 3, 3)
df_mental['Do you have Panic attack?'].value_counts().plot(
    kind='bar', color=['green', 'red'])
plt.title('Panic Attack')
plt.xlabel('')

plt.tight_layout()
plt.savefig(os.path.join(figures_path, 'mental_health_analysis.png'))
plt.show()
print("✅ Mental health chart saved!")
```

```python
# Explore Dropout Dataset
print("="*50)
print("🚨 DROPOUT DATASET")
print("="*50)
print(f"\nShape: {df_dropout.shape}")
print(f"\nColumns: {df_dropout.columns.tolist()}")
print(f"\nMissing Values:\n{df_dropout.isnull().sum()}")
print("\nTarget column value counts:")
print(df_dropout['Target'].value_counts())
print("\nFirst 3 rows:")
df_dropout.head(3)
```

```python
# Dropout Visualizations
plt.figure(figsize=(12, 4))

# Plot 1 - Target Distribution
plt.subplot(1, 3, 1)
df_dropout['Target'].value_counts().plot(
    kind='bar', color=['green', 'red', 'orange'])
plt.title('Student Status Distribution')
plt.xlabel('')

# Plot 2 - Age at Enrollment vs Target
plt.subplot(1, 3, 2)
sns.boxplot(x='Target', y='Age at enrollment', data=df_dropout)
plt.title('Age vs Student Status')

# Plot 3 - Scholarship vs Target
plt.subplot(1, 3, 3)
pd.crosstab(df_dropout['Target'], 
            df_dropout['Scholarship holder']).plot(
    kind='bar', color=['red', 'green'])
plt.title('Scholarship vs Status')
plt.xlabel('')

plt.tight_layout()
plt.savefig(os.path.join(figures_path, 'dropout_analysis.png'))
plt.show()
print("✅ Dropout chart saved!")

```

```python
# Key factors affecting final grade
plt.figure(figsize=(12, 8))

# Plot 1 - Internet access vs Grade
plt.subplot(2, 3, 1)
sns.boxplot(x='internet', y='G3', data=df_performance)
plt.title('Internet vs Grade')

# Plot 2 - Higher education goal vs Grade
plt.subplot(2, 3, 2)
sns.boxplot(x='higher', y='G3', data=df_performance)
plt.title('Wants Higher Education vs Grade')

# Plot 3 - Family support vs Grade
plt.subplot(2, 3, 3)
sns.boxplot(x='famsup', y='G3', data=df_performance)
plt.title('Family Support vs Grade')

# Plot 4 - Romantic relationship vs Grade
plt.subplot(2, 3, 4)
sns.boxplot(x='romantic', y='G3', data=df_performance)
plt.title('Romantic Relationship vs Grade')

# Plot 5 - Address vs Grade
plt.subplot(2, 3, 5)
sns.boxplot(x='address', y='G3', data=df_performance)
plt.title('Urban vs Rural vs Grade')

# Plot 6 - Gender vs Grade
plt.subplot(2, 3, 6)
sns.boxplot(x='sex', y='G3', data=df_performance)
plt.title('Gender vs Grade')

plt.tight_layout()
plt.savefig(os.path.join(figures_path, 'performance_factors.png'))
plt.show()
print("✅ Performance factors chart saved!")
```

```python
# Mental Health Pattern Analysis
plt.figure(figsize=(12, 8))

# Plot 1 - CGPA vs Depression
plt.subplot(2, 2, 1)
sns.countplot(x='What is your CGPA?', 
              hue='Do you have Depression?',
              data=df_mental)
plt.title('CGPA vs Depression')
plt.xticks(rotation=45)

# Plot 2 - Year of Study vs Anxiety
plt.subplot(2, 2, 2)
sns.countplot(x='Your current year of Study',
              hue='Do you have Anxiety?',
              data=df_mental)
plt.title('Year of Study vs Anxiety')
plt.xticks(rotation=45)

# Plot 3 - Gender vs Depression
plt.subplot(2, 2, 3)
sns.countplot(x='Choose your gender',
              hue='Do you have Depression?',
              data=df_mental)
plt.title('Gender vs Depression')

# Plot 4 - Seek Treatment
plt.subplot(2, 2, 4)
df_mental['Did you seek any specialist for a treatment?'].value_counts().plot(
    kind='pie', autopct='%1.1f%%',
    colors=['red', 'green'])
plt.title('Sought Treatment?')

plt.tight_layout()
plt.savefig(os.path.join(figures_path, 'mental_health_patterns.png'))
plt.show()
print("✅ Mental health patterns saved!")
```

```python
# Behavior Pattern Analysis
plt.figure(figsize=(12, 8))

# Plot 1 - Raised Hands vs Class
plt.subplot(2, 2, 1)
sns.boxplot(x='Class', y='raisedhands', data=df_behavior)
plt.title('Raised Hands vs Performance')

# Plot 2 - Visited Resources vs Class
plt.subplot(2, 2, 2)
sns.boxplot(x='Class', y='VisITedResources', data=df_behavior)
plt.title('Visited Resources vs Performance')

# Plot 3 - Absences vs Class
plt.subplot(2, 2, 3)
sns.countplot(x='StudentAbsenceDays', 
              hue='Class', data=df_behavior)
plt.title('Absences vs Performance')

# Plot 4 - Gender vs Class
plt.subplot(2, 2, 4)
sns.countplot(x='gender', 
              hue='Class', data=df_behavior)
plt.title('Gender vs Performance')

plt.tight_layout()
plt.savefig(os.path.join(figures_path, 'behavior_patterns.png'))
plt.show()
print("✅ Behavior patterns saved!")
```

```python
# Dropout Pattern Analysis
plt.figure(figsize=(12, 8))

# Plot 1 - Tuition fees vs Dropout
plt.subplot(2, 2, 1)
sns.countplot(x='Tuition fees up to date',
              hue='Target', data=df_dropout)
plt.title('Tuition Fees vs Status')

# Plot 2 - Scholarship vs Dropout
plt.subplot(2, 2, 2)
sns.countplot(x='Scholarship holder',
              hue='Target', data=df_dropout)
plt.title('Scholarship vs Status')

# Plot 3 - Age vs Dropout
plt.subplot(2, 2, 3)
sns.boxplot(x='Target', 
            y='Age at enrollment',
            data=df_dropout)
plt.title('Age vs Status')

# Plot 4 - Gender vs Dropout
plt.subplot(2, 2, 4)
sns.countplot(x='Gender',
              hue='Target', data=df_dropout)
plt.title('Gender vs Status')

plt.tight_layout()
plt.savefig(os.path.join(figures_path, 'dropout_patterns.png'))
plt.show()
print("✅ Dropout patterns saved!")
```
