```python
# Visualize Grade Distribution
import os

# Create figures directory if not exists
figures_path = r'C:\Users\Amsha\Desktop\EduMindAI\reports\figures'
os.makedirs(figures_path, exist_ok=True)

plt.figure(figsize=(12, 4))

# Plot 1 - Final Grade Distribution
plt.subplot(1, 3, 1)
sns.histplot(df_performance['G3'], bins=20, color='blue')
plt.title('Final Grade Distribution')
plt.xlabel('Grade (G3)')

# Plot 2 - Study Time vs Grade
plt.subplot(1, 3, 2)
sns.boxplot(x='studytime', y='G3', data=df_performance, color='green')
plt.title('Study Time vs Grade')

# Plot 3 - Failures vs Grade
plt.subplot(1, 3, 3)
sns.boxplot(x='failures', y='G3', data=df_performance, color='red')
plt.title('Failures vs Grade')

plt.tight_layout()
plt.savefig(os.path.join(figures_path, 'performance_analysis.png'))
plt.show()
print("✅ Chart saved!")
```

```python
# Explore Mental Health Dataset
print("="*50)
print("😰 MENTAL HEALTH DATASET")
print("="*50)
print(f"\nShape: {df_mental.shape}")
print(f"\nColumns: {df_mental.columns.tolist()}")
print(f"\nMissing Values:\n{df_mental.isnull().sum()}")
print("\nFirst 5 rows:")
df_mental.head()
```

```python
# Mental Health Visualizations
plt.figure(figsize=(12, 4))

# Plot 1 - Depression
plt.subplot(1, 3, 1)
df_mental['Do you have Depression?'].value_counts().plot(
    kind='bar', color=['green', 'red'])
plt.title('Depression')
plt.xlabel('')

# Plot 2 - Anxiety
plt.subplot(1, 3, 2)
df_mental['Do you have Anxiety?'].value_counts().plot(
    kind='bar', color=['green', 'red'])
plt.title('Anxiety')
plt.xlabel('')

# Plot 3 - Panic Attack
plt.subplot(1, 3, 3)
df_mental['Do you have Panic attack?'].value_counts().plot(
    kind='bar', color=['green', 'red'])
plt.title('Panic Attack')
plt.xlabel('')

plt.tight_layout()
plt.savefig(os.path.join(figures_path, 'mental_health_analysis.png'))
plt.show()
print("✅ Mental health chart saved!")
```

```python
# Explore Dropout Dataset
print("="*50)
print("🚨 DROPOUT DATASET")
print("="*50)
print(f"\nShape: {df_dropout.shape}")
print(f"\nColumns: {df_dropout.columns.tolist()}")
print(f"\nMissing Values:\n{df_dropout.isnull().sum()}")
print("\nTarget column value counts:")
print(df_dropout['Target'].value_counts())
print("\nFirst 3 rows:")
df_dropout.head(3)
```

```python
# Dropout Visualizations
plt.figure(figsize=(12, 4))

# Plot 1 - Target Distribution
plt.subplot(1, 3, 1)
df_dropout['Target'].value_counts().plot(
    kind='bar', color=['green', 'red', 'orange'])
plt.title('Student Status Distribution')
plt.xlabel('')

# Plot 2 - Age at Enrollment vs Target
plt.subplot(1, 3, 2)
sns.boxplot(x='Target', y='Age at enrollment', data=df_dropout)
plt.title('Age vs Student Status')

# Plot 3 - Scholarship vs Target
plt.subplot(1, 3, 3)
pd.crosstab(df_dropout['Target'], 
            df_dropout['Scholarship holder']).plot(
    kind='bar', color=['red', 'green'])
plt.title('Scholarship vs Status')
plt.xlabel('')

plt.tight_layout()
plt.savefig(os.path.join(figures_path, 'dropout_analysis.png'))
plt.show()
print("✅ Dropout chart saved!")

```
